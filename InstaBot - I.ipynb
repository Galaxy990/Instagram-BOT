### Q 5) Like / Unlike posts :-

    i) Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
    ii) Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
driv.find_element_by_xpath('//input[@placeholder = "Search"]').send_keys('dilsefoodie')
WebDriverWait(driv,10).until(EC.presence_of_element_located((By.XPATH,'//div[@class = "fuqBx "]//div[@class = "_7UhW9   xLCgt       qyrsm KV-D4          uL8Hv         "]'))).click()
a = WebDriverWait(driv,10).until(EC.presence_of_element_located((By.CLASS_NAME,'_9AhH0')))
a.click()

def liking():
    
    for i in range(30):
        a = WebDriverWait(driv,10).until(EC.presence_of_element_located((By.CLASS_NAME,'fr66n')))
        b = WebDriverWait(driv,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.QBdPU  span svg')))
        txt = b.get_attribute('aria-label')

        if txt == 'Like':
            a.click()
            time.sleep(1)

        elif txt == 'Unlike':
            print(f"Post No {i + 1} is Already Liked")

        driv.find_element_by_xpath('//a[@class = " _65Bje  coreSpriteRightPaginationArrow"]').click()
        time.sleep(3)
    
def unliking():
    
    for i in range(30):
        a = WebDriverWait(driv,10).until(EC.presence_of_element_located((By.CLASS_NAME,'fr66n')))
        b = WebDriverWait(driv,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.QBdPU  span svg')))
        txt = b.get_attribute('aria-label')

        if txt == 'Like':
            print(f"Post No {i + 1} is Already Unliked")

        elif txt == 'Unlike':
            a.click()
            time.sleep(1)

        driv.find_element_by_xpath('//a[@class = " _65Bje  coreSpriteRightPaginationArrow"]').click()
        time.sleep(3)
    
def automated_liking_unliking():
    print("Press 1 to 'Like' the Top 30 posts of 'dilsefoodie'\nPress 2 to 'Unlike' the Top 30 posts of 'dilsefoodie'")
    x = input()
    print()
    if x == '1':
        liking()
    else:
        unliking()
        
automated_liking_unliking()
```

#### Usernames of First 500 followers of "foodtalkindia"

```python
driv.find_element_by_xpath('//input[@placeholder = "Search"]').send_keys('foodtalkindia')
WebDriverWait(driv,100).until(EC.presence_of_element_located((By.XPATH,'//div[@class = "fuqBx "]//div[@class = "_7UhW9   xLCgt       qyrsm KV-D4          uL8Hv         "]'))).click()
time.sleep(3)
driv.find_element_by_xpath('//a[@class = "-nal3 "]').click()
a = WebDriverWait(driv,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,'div.isgrP')))
c_ht = driv.execute_script('return arguments[0].scrollHeight;',a)  # Finding the "scollHeight" of the pop-up list of followers which is present in the UNORDERED LIST :- <ul class = 'jSC57  _6xe7A' and the unordered list in turn is present in <div class = "isgrP">
i = 0

while i < 42:
    driv.execute_script('arguments[0].scrollTo(0,arguments[1]);',a,c_ht)  # Scrolling down the pop-up window until new elements get loaded into the DOM.
    time.sleep(3)
    c_ht = driv.execute_script('return arguments[0].scrollHeight;',a)  # Updating the height because after new elements are loaded the height of the unordered list increases.
    i += 1

x = driv.find_elements_by_xpath('//a[@class = "FPmhX notranslate  _0imsa "]')  # finding the <a> anchor tag which contains the username of an user 
y = driv.find_elements_by_css_selector('div.Pkbci button')    # Finding the usernames which we are following but those who don't follow us, We include only those usernames on which the follow status is "Following" and not "Requested",  NOTE :- It will be used in the 2nd sub question
x = x[:500]
print("Usernames Of First 500 followers of 'foodtalkindia' are :-\n")
print(f"Length Of the followers list is :- {len(x)}\n")
for i in x:
    print(i.text)
```

### Q 7) Check the story of ‘coding.ninjas’. Consider the following Scenarios and print messages accordingly

        i) If You have already seen the story.
        ii) Or The user has no story.
        iii) Or View the story if not yet seen.

```python
x = driv.find_elements_by_css_selector('div.Fd_fQ')  # Searching for all the current stories
y = driv.find_elements_by_css_selector('div.Fd_fQ button')  # Searching for the <tag> which contains the status of story, whether it is "seen" or "not seen"
cnt = -1

for i in range(len(x)):
    txt = y[i].get_attribute('aria-label')
    if 'coding.ninjas' in txt:  # If story name contains conding.ninjas in it then we further check whether story is "seen" or "not seen"
        cnt = 0
        seen_or_not = txt[txt.index(',') + 2:]
        if seen_or_not == 'seen':  # If story is Already Seen print a message
            print('Story is Already Seen')
        elif seen_or_not == 'not seen':  # If Story is not seen then print a message and proceed to view the story
            print('Story is not seen, hence it will be viewed now')
            x[i].click()
            
if cnt == -1:  # If story name did not contain "coding.ninjas" in it, then we know that currently there is no story by "coding.ninjas", hence we print a message
    print('Currently there is no story posted by coding-ninjas')
```
